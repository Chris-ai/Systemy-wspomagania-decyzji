```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('INCOME.TXT', sep ='\t', decimal=',')
data.head()
```

```python
numberOfRows = len(data)
percentage = 0.1
n = int(numberOfRows*percentage)
data.nsmallest(n, 'Aktywa')
```

```python
columns = ['Aktywa', 'Przych']
data[columns]

len(columns)
```

```python
data['Hrabstwo'].value_counts()

type(data.describe())
```

```python
# data['Hrabstwo'] = pd.factorize(data['Hrabstwo'])[0] #Kolejnosc występowania
data['Hrabstwo_num_kol_wystp'] = pd.factorize(data['Hrabstwo'])[0]
# data['Hrabstwo'] = pd.Categorical(data['Hrabstwo']).codes #alfabetycznie
data['Hrabstwo_num_alf'] = pd.Categorical(data['Hrabstwo']).codes

data.head(15)
```

```python
# #q -> liczba przedziałów
data['Aktywa_przedzial'] = pd.qcut(data['Aktywa'], q=5)
data['Przych_przedzial'] = pd.qcut(data['Przych'], q=5)

data.head(20)
```

```python
#Normalizacja
data['Aktywa_norm'] = ( ((data['Aktywa'])  - (data['Aktywa']).mean())/data['Aktywa'].std())
data['Przych_norm'] = ( ((data['Przych'])  - (data['Przych']).mean())/data['Przych'].std())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
def savePlotAsPng(data, name):
    fig = plt.gcf()
    fig.set_size_inches(25.5, 18.5)

    ax = sns.lineplot(data= data, x = 'k', y='acc_score', hue='metric', style='metric', palette=sns.color_palette("husl",4), markers=['o','o','o','o'])

incomeData = pd.read_csv('data/data-income-quality-assessment.txt', sep=' ', decimal=',', names=['k', 'acc_score', 'metric'])
incomeData['acc_score'] = incomeData['acc_score'].astype(float)

irisData = pd.read_csv('data/data-iris-quality-assessment.txt', sep=' ', decimal=',', names=['k', 'acc_score', 'metric'])
irisData['acc_score'] = irisData['acc_score'].astype(float)

incomeNormalizedData = pd.read_csv('data/data-incomeNormalized-quality-assessment.txt', sep=' ', decimal=',', names=['k', 'acc_score', 'metric'])
incomeNormalizedData['acc_score'] = incomeNormalizedData['acc_score'].astype(float)

irisNormalizedData = pd.read_csv('data/data-irisNormalized-quality-assessment.txt', sep=' ', decimal=',', names=['k', 'acc_score', 'metric'])
irisNormalizedData['acc_score'] = irisNormalizedData['acc_score'].astype(float)
```

```python
savePlotAsPng(incomeData, 'income')
```

```python
savePlotAsPng(irisData, 'iris')
```

```python
savePlotAsPng(incomeNormalizedData, 'incomeNormalized')
```

```python
savePlotAsPng(irisNormalizedData, 'irisNormalized')
```
